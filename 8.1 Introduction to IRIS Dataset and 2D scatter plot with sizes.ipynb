```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sklearn

#Load iris.csv into pandas dataframe
iris = pd.read_csv('C:\Austine\Data Science\Applied AI\Exercises\Iris dataset\iris.csv')
```

```python
#Print the number of data point and features
print(iris.shape)
```

```python
#What are the column names in our dataset?
print (iris.columns)
```

```python
#How many data points for each class are present?
#How many flowers of each species are present?

iris['species'].value_counts()
#This is a balanced dataset as we have equal number of datapoints for each species
```

```python
#2D scatter plots
#Always understand the axis: labels and scale

iris.plot(kind='scatter', x='sepal_length', y='sepal_width');
plt.show()

#We cannot make much sense out of it
#What if we color the points by class label / flower type
```

```python
#2-D scatter plot with color-coding for each flower type/class
#Here 'sns' corresponds to seaborn

sns.set_style("whitegrid");
sns.FacetGrid(iris, hue = "species", size = 4) \
    .map(plt.scatter, "sepal_length", "sepal_width") \
    .add_legend();
plt.show();

#We can notice here that the blue dots can be easily separated from read and green by drawing a line
#But red and green data points can not be easily separated
#Can we draw multiple 2-D scatter plots for each combination of features?
#How many combinations exists? 4C2 = 6
```

```python
#We can go further and also add size to our plot
#The plot will not show the datapoints with the size of the defined column
```

```python
iris = np.array(iris)
new = np.copy(iris)
new[iris=='setosa'] = 1
new[iris=='versicolor'] = 2
new[iris=='virginica'] = 3
new = new.T
new
```

```python
size = 100*new[3]
size = size.astype(int)

plt.figure(figsize=(10,6))
plt.scatter(new[0], new[1], alpha=0.8, s=size,
             c = new[4], cmap='viridis')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width');

plt.colorbar()
plt.show();
```

```python
#Observations
#1. Using sepal_length and sepal_width, we can distinguish Setosa flowers from others
#2. Seperating Versicolor from Virginica is much harder as they have considerable overlap
```
